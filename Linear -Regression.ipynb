```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python

data = pd.read_csv("/content/drive/MyDrive/MSC_MU/MSC_SEM_2/ML_ALL_PROGRAM_DS_SEM_2/25-Mar/cputime.csv")
```

```python
x = data.iloc[:, 0:1]
y = data.iloc[:, 1]
```

```python
print("First 5 rows of x (Disk I/O):")
print(x.head())
print("\nFirst 5 rows of y (CPU time):")
print(y.head())
```

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=0)
```

```python
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
y_pred_train = model.predict(X_train)
y_pred_test = model.predict(X_test)
y_pred_40 = model.predict([[40]])
```

```python
print(f"\nValue of CPU time for disk I/O = 40 is {y_pred_40[0]:.2f}")
```

```python
plt.scatter(X_train, y_train, color='lightcoral', label='Training data')
plt.plot(X_train, y_pred_train, color='firebrick', label='Regression line')
plt.title('CPU time vs Disk I/O')
plt.xlabel('Disk I/O')
plt.ylabel('CPU time')
plt.legend()
plt.show()
```

#Profit vs Population



```python
from sklearn import metrics

data = pd.read_csv("/content/drive/MyDrive/MSC_MU/MSC_SEM_2/ML_ALL_PROGRAM_DS_SEM_2/25-Mar/linearregressiondataset.csv")
```

```python

x = data.iloc[:, 0:1]
y = data.iloc[:, 1]
```

```python


print("First 5 rows of x (Population):")
print(x.head())
print("\nFirst 5 rows of y (Profit):")
print(y.head())
```

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=0)
```

```python
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
print(f"\nCoefficient: {model.coef_[0]:.4f}")
print(f"Intercept: {model.intercept_:.4f}")

# Predictions
y_pred_train = model.predict(X_train)
y_pred_test = model.predict(X_test)
y_pred_20_27 = model.predict([[20.27]])
```

```python
print(f"\nPredicted profit for a population of 20.27 lakh: ${y_pred_20_27[0]:,.2f}")

print("\nModel Evaluation on Test Set:")
print(f"Mean Absolute Error (MAE): ${metrics.mean_absolute_error(y_test, y_pred_test):,.2f}")
print(f"Mean Squared Error (MSE): ${metrics.mean_squared_error(y_test, y_pred_test):,.2f}")
print(f"Root Mean Squared Error (RMSE): ${np.sqrt(metrics.mean_squared_error(y_test, y_pred_test)):,.2f}")
```

```python
plt.scatter(X_train, y_train, color='lightcoral', label='Training data')
plt.plot(X_train, y_pred_train, color='firebrick', label='Regression line')
plt.title('Profit vs Population')
plt.xlabel('Population (lakhs)')
plt.ylabel('Profit ($)')
plt.legend()
plt.show()
```
